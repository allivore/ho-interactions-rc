```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open(
f"Results/Evaluation_Data/fhn_d005-o3n-monte_carlo_lorenz300n_dt0.01-35/" \
f"RNN-esn2_auto-RDIM_3-N_used_10000-SIZE_300-D_10-RADIUS_1.0-SIGMA_1-DL_2000-NL_0-IPL_6000-REG_0.0001-WID_136/" \
"results.pickle",
"rb") as infile:
    data = pickle.load(infile)
pred_x = data['predictions_all_TEST'][0, :, 0]
pred_y = data['predictions_all_TEST'][0, :, 1]
pred_z = data['predictions_all_TEST'][0, :, 2]

true_x = data['truths_all_TEST'][0, :, 0]
true_y = data['truths_all_TEST'][0, :, 1]
true_z = data['truths_all_TEST'][0, :, 2]
```

```python
t = np.linspace(0, 60, 6000)
fig, axs = plt.subplots(3, 1, sharex=True, figsize=[10, 5])

axs[0].plot(t[:2500], pred_x[:2500], label="Predicted", color="red")
axs[0].plot(t[:2500], true_x[:2500], label="Target", color="teal")
axs[0].legend(loc="upper left")

axs[1].plot(t[:2500], pred_y[:2500], label="Predicted", color="red")
axs[1].plot(t[:2500], true_y[:2500], label="Target", color="teal")
axs[1].legend(loc="upper left")

axs[2].plot(t[:2500], pred_z[:2500], label="Predicted", color="red")
axs[2].plot(t[:2500], true_z[:2500], label="Target", color="teal")
axs[2].legend(loc="upper left")
```

```python
ax = plt.figure(figsize=(10, 10)).add_subplot(projection='3d')
ax.plot(true_x, true_y, true_z, color="teal")
```

```python
ax = plt.figure(figsize=(10, 10)).add_subplot(projection='3d')
ax.plot(pred_x, pred_y, pred_z, color="red")
```
