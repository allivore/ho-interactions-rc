```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
with open(f"Results/Evaluation_Data/fhn_d005-o3n-a=1.0-r500-opt-fix/" \
        f"RNN-esn2_auto-RDIM_3-N_used_30000-SIZE_500-" \
        f"D_10-RADIUS_1.9-SIGMA_1-DL_10000-NL_0-IPL_20000-" \
        f"REG_0.0001-WID_7/results.pickle", "rb") as infile:
    data = pickle.load(infile)
pred = data['predictions_all_TEST'][0, :, 0]
true = data['truths_all_TEST'][0, :, 0]
rmse = mean_squared_error(true, pred, squared=False)
print(rmse)
```

```python
fig = plt.figure(figsize=(10, 2.5))
plt.plot(true, label="Target", color="teal")
plt.legend(loc="upper left")
```

```python
fig = plt.figure(figsize=(10, 2.5))
plt.plot(pred, label="All simplices", color="red")
plt.legend(loc="upper left")
```
