```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
with open(f"Results/Evaluation_Data/fhn_d005-o2n-a=1.0n0.0-r500-opt-seed=1/" \
        f"RNN-esn2_auto-RDIM_3-N_used_30000-SIZE_500-" \
        f"D_16-RADIUS_0.9-SIGMA_1-DL_10000-NL_0-IPL_20000-" \
        f"REG_0.0001-WID_1/results.pickle", "rb") as infile:
    data = pickle.load(infile)
pred = data['predictions_all_TEST'][0, :, 0]
true = data['truths_all_TEST'][0, :, 0]
rmse = mean_squared_error(true, pred, squared=False)
print(rmse)


fig, axs = plt.subplots(2, 1, sharex=True, figsize=[10, 5])

axs[0].plot(pred, label="Predicted", color="red")
axs[0].plot(true, label="Target", color="teal")
axs[0].legend(loc="upper left")

axs[1].plot(pred-true, label="Difference", color="orange")
axs[1].legend(loc="upper left")

fig.suptitle("First Order")
```
