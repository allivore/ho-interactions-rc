```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open(
f"Results/Evaluation_Data/fhn_d005-o3n-monte_carlo_rossler300n_dt0.02-9/" \
f"RNN-esn2_auto-RDIM_3-N_used_26000-SIZE_300-D_10-RADIUS_1.0-SIGMA_1-DL_1000-NL_0-IPL_25000-REG_0.0001-WID_495/" \
"results.pickle",
"rb") as infile:
    data = pickle.load(infile)
pred_x = data['predictions_all_TEST'][0, :, 0]
pred_y = data['predictions_all_TEST'][0, :, 1]
pred_z = data['predictions_all_TEST'][0, :, 2]

true_x = data['truths_all_TEST'][0, :, 0]
true_y = data['truths_all_TEST'][0, :, 1]
true_z = data['truths_all_TEST'][0, :, 2]
```

```python
t = np.linspace(0, 500, 25000)
fig, axs = plt.subplots(3, 1, sharex=True, figsize=[10, 5])

axs[0].plot(t[:15000], pred_x[:15000], label="Predicted", color="red")
axs[0].plot(t[:15000], true_x[:15000], label="Target", color="teal")
axs[0].legend(loc="upper left")

axs[1].plot(t[:15000], pred_y[:15000], label="Predicted", color="red")
axs[1].plot(t[:15000], true_y[:15000], label="Target", color="teal")
axs[1].legend(loc="upper left")

axs[2].plot(t[:15000], np.exp(pred_z)[:15000], label="Predicted", color="red")
axs[2].plot(t[:15000], np.exp(true_z)[:15000], label="Target", color="teal")
axs[2].legend(loc="upper left")
```

```python
ax = plt.figure(figsize=(10, 10)).add_subplot(projection='3d')
ax.plot(true_x, true_y, np.exp(true_z), color="teal")
```

```python
ax = plt.figure(figsize=(10, 10)).add_subplot(projection='3d')
ax.plot(pred_x, pred_y, np.exp(pred_z), color="red")
```
