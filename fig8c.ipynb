```python
import pickle
import numpy as np 
import matplotlib.pyplot as plt
```

```python
with open(
f"Results/Evaluation_Data/fhn_d005-o3n-monte_carlo_eeg-13/" \
f"RNN-esn2_auto-RDIM_6-N_used_60000-SIZE_300-D_10-RADIUS_1.0-SIGMA_1-DL_1000-NL_0-IPL_59000-REG_0.0001-WID_217/" \
"results.pickle",
"rb") as infile:
    data = pickle.load(infile)
pred = data['predictions_all_TEST'][0, :, 0]
true = data['truths_all_TEST'][0, :, 0]
```

```python
t = np.linspace(0, int(59000*0.001), 59000)
fig, axs = plt.subplots(1, 1, figsize=(10, 6))
axs.plot(t[:1000], true[:1000], color="teal")
```

```python
t = np.linspace(0, int(59000*0.001), 59000)
fig, axs = plt.subplots(1, 1, figsize=(10, 6))
axs.plot(t[:1000], pred[:1000], color="red")
```

```python
t = np.linspace(0, int(59000*0.001), 59000)
fig, axs = plt.subplots(1, 1, figsize=(10, 6))
axs.plot(t[:1000], (pred[:1000]-true[:1000])/np.std(true[:1000]), color="orange")
```
